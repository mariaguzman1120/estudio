### Rango intercuartil

```python
iqr_ = iqr(df_imputacion_simple, axis=0, rng=(25, 75), interpolation='midpoint')
```

```python
iqr_
```

```python
q1 = np.percentile(df_imputacion_simple, 25, axis=0, interpolation='midpoint')
q3 = np.percentile(df_imputacion_simple, 75, axis=0, interpolation='midpoint')

min_limite = q1 - 1.5 * iqr_
max_limite = q3 + 1.5 * iqr_

print('Límites inferiores = ', min_limite)
print("\n")
print('Límites superiores = ', max_limite)
```

```python
outliers = []
values = []
columnas = df_imputacion_simple.columns
for i in range(len(columnas)):
    serie = df_imputacion_simple[columnas[i]]
    min_ = min_limite[i]
    max_ = max_limite[i]
    contador = 0
    lista = []
    for value in serie.unique():
        if value < min_ or value > max_:
            contador += 1
            d = df_imputacion_simple[serie == value].index
            for j in d:
                lista.append(j)
    outliers.append(contador)
    values.append(lista)
```

```python
dic2 = {
    'Columnas': df_imputacion_simple.columns,
    'Outliers': outliers
}

tabla_outliers = pd.DataFrame(dic2)
```

```python
tabla_outliers
```

```python
set_ = set(values[0])
for m in range(1, len(values)):
    set_ = set_.intersection(set(values[m]))
    
len(set_)
```

```python
r = []
valor_p = []
data = df_nueva.values
data = np.asarray(data)

for n in range(data.shape[1]):
    for m in range(data.shape[1]):
        r_valor, p_valor = pearsonr(data[:, n], data[:, m])
        r.append(r_valor)
        valor_p.append(p_valor)
```

```python
r = np.asarray(r)
valor_p = np.asarray(valor_p)

pearson_r = r.reshape(data.shape[1], data.shape[1])
p_value = valor_p.reshape(data.shape[1], data.shape[1])

plt.figure(figsize=(20, 20))
ax = sns.heatmap(
    pearson_r, 
    annot=True, 
    fmt='g', 
    xticklabels=df_nueva.columns, 
    yticklabels=df_nueva.columns,
    annot_kws={'size': 7.5})

plt.show()
```

```python
nombres = ["beta", "gamma", "expon", "chi2", "norm", "t", "uniform"]
ks_valor = []
p_valor = []
data = df_imputacion_simple.values

for distribuciones in nombres:
    dis = getattr(scipy.stats, distribuciones)
    for n in df_imputacion_simple.columns:
        A = np.asarray(df_imputacion_simple[n])
        par = dis.fit(A)
        D, p = scipy.stats.kstest(
            df_imputacion_simple[n],
            distribuciones, 
            N=100, 
            mode="auto", 
            args=par)
        pp = np.around(p, 10) 
        DD = np.around(D, 5) 
        p_valor.append(pp)
        ks_valor.append(DD)

ks_valor = np.asarray(ks_valor)
p_valor = np.asarray(p_valor)

ks_valor = ks_valor.reshape(len(nombres), df_imputacion_simple.shape[1])

p_valor = p_valor.reshape(len(nombres), df_imputacion_simple.shape[1])

ks_valor = pd.DataFrame(ks_valor, index=nombres, columns=df_imputacion_simple.columns)
p_valor = pd.DataFrame(p_valor, index=nombres, columns=df_imputacion_simple.columns)
```
