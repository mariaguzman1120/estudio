# Top 5 de canales por categoría con más vistas

```python
categories = df.set_index('Youtube Channel').groupby('Category')['Video Views'].nlargest(5)
```

```python
categories = pd.DataFrame(categories)
categories.reset_index(inplace=True)
categories
```

```python
categories['Youtube Channel'] = categories['Youtube Channel'].apply(lambda x: textwrap.fill(x, 20))
```

```python
fig, axes = plt.subplots(3, 6, figsize=(16,8))

for cat, idx in zip(categories['Category'].unique(), axes.ravel()):
    
    df = categories[categories['Category'] == cat]
    g = sns.barplot(
        ax=idx,
        data=df, 
        x='Youtube Channel', 
        y='Video Views')

    g.set(yticklabels=[])
    g.set(xlabel=None)
    g.set(ylabel=None)
    g.set(title=f'{cat}')
    idx.tick_params(axis='x', rotation=80)


plt.subplots_adjust(
    left=0.1, 
    bottom=0.1, 
    right=0.9, 
    top=0.9,
    wspace=0.2,
    hspace=2.5) 

plt.show()
```

```python
fig, axes = plt.subplots(3, 6, figsize=(20,10))

for cat, idx in zip(categories['Category'].unique(), axes.ravel()):
    
    df = categories[categories['Category'] == cat]
    g = df.plot(
        ax=idx,
        kind='pie',
        x='Youtube Channel', 
        y='Video Views',
        autopct='%1.1f%%', 
        startangle=90,
        shadow=True,
        labels=None,
        pctdistance=1.12, ylabel=None, title=f'{cat}', yticks=[], legend=False)
    g.set_ylabel('')
    g.legend(labels=df['Youtube Channel'], loc='best', bbox_to_anchor=(1, 0), fontsize=7)
    
plt.subplots_adjust(
    left=0.1, 
    bottom=0.1, 
    right=0.9, 
    top=0.9,
    wspace=0.5,
    hspace=1.5) 

plt.show()
```
